```python
import pennylane as qp
```

```python
dev = qp.device("default.qubit")
```

```python
#Bell State formation
def circuit():
    qp.H(wires=0)
    qp.CNOT(wires=[0,1])
    return qp.probs()

qp.draw_mpl(circuit)();
```

```python
@qp.qnode(dev)
def circuit2():
    qp.RX(0.1, wires=0)
    qp.RY(0.2, wires=0)
    return qp.probs()

qp.draw_mpl(circuit2)()
circuit2()
```

```python
#Adding backend device to existing circuit
circuit3 = qp.QNode(circuit,dev)

circuit3()
```

```python
# Template Circuits

@qp.qnode(dev)
def circuit4():
    qp.BasisEmbedding([1,0,1], wires=[0,1,2])
    return qp.
```
